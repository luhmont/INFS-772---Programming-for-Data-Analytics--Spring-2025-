```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
companysales = pd.read_csv('company_sales_data.csv')
companysales.head()
```

```python
#Exercise 1: Read Total profit of all months and show it using a line plot

x = companysales['month_number']
y = companysales['total_profit']

plt.plot(x,y)
plt.xlabel("Month number")
plt.xticks(np.arange(1,13,1))
plt.ylabel("Total Profit")
plt.yticks(np.arange(100000,600000,100000))
plt.title("Company profit per month")
plt.show()
```

```python
#Exercise 2: Get Total profit of all months and show line plot with the following Style properties
#Generated line plot must include following Style properties: –
# Line Style dotted and Line-color should be red
# Show legend at the lower right location.
# X label name = Month Number
# Y label name = Sold units number
# Add a circle marker.
# Line marker color as read
# Line width should be 3

plt.plot(x,y, linestyle='--', color='r', marker='o', linewidth=3, label="Profit data of last year", markerfacecolor='k')
plt.legend(loc='lower right')
plt.xlabel("Month number")
plt.xticks(np.arange(1,13,1))
plt.ylabel("Sold units number")
plt.yticks(np.arange(100000,600000,100000))
plt.title("Company profit per month")
plt.show()
```

```python
#Exercise 3: Read all product sales data and show it using a multiline plot


y = companysales['bathingsoap']
plt.plot(x,y, color='r', marker='o', linewidth=3, label='Bathing soap sales data')
plt.legend(loc='upper left')
y = companysales['facecream']
plt.plot(x,y, color='b', marker='o', linewidth=3, label='Facecream sales data')
plt.legend(loc='upper left')
y = companysales['facewash']
plt.plot(x,y, color='g', marker='o', linewidth=3, label='Facewash sales data')
plt.legend(loc='upper left')
y = companysales['moisturizer']
plt.plot(x,y, color='c', marker='o', linewidth=3, label='Moisturizer sales data')
plt.legend(loc='upper left')
y = companysales['shampoo']
plt.plot(x,y, color='m', marker='o', linewidth=3, label='Shampoo sales data')
plt.legend(loc='upper left')
y = companysales['toothpaste']
plt.plot(x,y, color='y', marker='o', linewidth=3, label='Toothpaste sales data')
plt.legend(loc='upper left')
plt.xlabel("Month number")
plt.xticks(np.arange(1,13,1))
plt.ylabel("Sold units number")
plt.yticks(np.arange(0, 20000,2000))
```

```python
#Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot

x = companysales['month_number']
y = companysales['toothpaste']

plt.scatter(x,y, color='b', marker='o', label="Tooth paste Sales data")
plt.grid(True, linestyle='--')
plt.xticks(np.arange(1,13,1))
plt.xlabel("Month Number")
plt.legend()

plt.show()
```

```python
#Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

y = companysales['facecream']
plt.bar(x-0.25/2,y, width=.25, color='b', label="Face Cream sales data")
plt.legend(loc='upper left')
y = companysales['facewash']
plt.bar(x+.25/2,y, width=.25, color='r', label="Face Wash sales data")
plt.legend(loc='upper left')
plt.xticks(np.arange(1,13,1))
plt.grid(True, linestyle='--')
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.show
```
