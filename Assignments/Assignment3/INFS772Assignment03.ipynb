```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
companysales = pd.read_csv('company_sales_data.csv')
companysales.head()
```

```python
#Exercise 1: Read Total profit of all months and show it using a line plot

x = companysales['month_number']
y = companysales['total_profit']

plt.plot(x,y)
plt.xlabel("Month number")
plt.xticks(np.arange(1,13,1))
plt.ylabel("Total Profit")
plt.yticks(np.arange(100000,600000,100000))
plt.title("Company profit per month")
plt.show()
```

```python
#Exercise 2: Get Total profit of all months and show line plot with the following Style properties
#Generated line plot must include following Style properties: 
# Line Style dotted and Line-color should be red
# Show legend at the lower right location.
# X label name = Month Number
# Y label name = Sold units number
# Add a circle marker.
# Line marker color as read
# Line width should be 3

plt.plot(x,y, linestyle='--', color='r', marker='o', linewidth=3, label="Profit data of last year", markerfacecolor='k')
plt.legend(loc='lower right')
plt.xlabel("Month number")
plt.xticks(np.arange(1,13,1))
plt.ylabel("Sold units number")
plt.yticks(np.arange(100000,600000,100000))
plt.title("Company profit per month")
plt.show()
```

```python
#Exercise 3: Read all product sales data and show it using a multiline plot


y = companysales['bathingsoap']
plt.plot(x,y, color='r', marker='o', linewidth=3, label='Bathing soap sales data')
plt.legend(loc='upper left')
y = companysales['facecream']
plt.plot(x,y, color='b', marker='o', linewidth=3, label='Facecream sales data')
plt.legend(loc='upper left')
y = companysales['facewash']
plt.plot(x,y, color='g', marker='o', linewidth=3, label='Facewash sales data')
plt.legend(loc='upper left')
y = companysales['moisturizer']
plt.plot(x,y, color='c', marker='o', linewidth=3, label='Moisturizer sales data')
plt.legend(loc='upper left')
y = companysales['shampoo']
plt.plot(x,y, color='m', marker='o', linewidth=3, label='Shampoo sales data')
plt.legend(loc='upper left')
y = companysales['toothpaste']
plt.plot(x,y, color='y', marker='o', linewidth=3, label='Toothpaste sales data')
plt.legend(loc='upper left')
plt.xlabel("Month number")
plt.xticks(np.arange(1,13,1))
plt.ylabel("Sold units number")
plt.yticks(np.arange(0, 20000,2000))
```

```python
#Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot

x = companysales['month_number']
y = companysales['toothpaste']

plt.scatter(x,y, color='b', marker='o', label="Tooth paste Sales data")
plt.grid(True, linestyle='--')
plt.xticks(np.arange(1,13,1))
plt.xlabel("Month Number")
plt.legend()

plt.show()
```

```python
#Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

y = companysales['facecream']
plt.bar(x-0.25/2,y, width=.25, color='b', label="Face Cream sales data")
plt.legend(loc='upper left')
y = companysales['facewash']
plt.bar(x+.25/2,y, width=.25, color='r', label="Face Wash sales data")
plt.legend(loc='upper left')
plt.xticks(np.arange(1,13,1))
plt.grid(True, linestyle='--')
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.show
```

```python
#Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk
#plt.plot(x, y)

y = companysales['bathingsoap']
plt.bar(x,y, color='b')
plt.grid(True, linestyle='--')
plt.title("bathingsoap sales data")
plt.xlabel("Month Number")
plt.ylabel("Sales units in number")
plt.xticks(np.arange(1,13,1))
plt.savefig("line_plot.png", dpi=300)  # Saves as PNG with high resolution
```

```python
#Exercise 7: Read the total profit of each month and show it using the histogram to see most common profit ranges

plt.hist(companysales['total_profit'], bins=6, color='b', label="Profit data")
plt.xlabel("profit range in dollar")
plt.ylabel('Actual Profit in dollar')
plt.title("Profit data")
plt.legend(loc='upper left')
plt.xticks(np.arange(150000,400000,25000))
```

```python
#Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart

totals = companysales[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']].sum()
plt.figure(figsize=(8,8))
plt.pie(totals, labels=totals.index, autopct='%1.1f%%', startangle=140)
```

```python
#Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

x = companysales['month_number']
y = companysales['total_units']

fig, axes = plt.subplots(2, 1, figsize=(12, 5))

# Bathing Soap Sales Plot
axes[0].plot(companysales['month_number'], companysales['bathingsoap'], marker='o', color='k', linestyle='-')
axes[0].set_title("Sales data of Bathingsoap")
axes[0].grid(True)


# Face Wash Sales Plot
axes[1].plot(companysales['month_number'], companysales['facewash'], marker='o', color='r', linestyle='-')
axes[1].set_title("Sales data of a facewash")
axes[1].set_xlabel("Month number")
axes[1].set_ylabel("Sales units in number")
axes[1].grid(True)
```

```python
#Exercise 10: Read all product sales data and show it using the stack plot

x = companysales['month_number']
plt.stackplot(x, companysales['facecream'], companysales['facewash'], companysales['toothpaste'], companysales['bathingsoap'],companysales['shampoo'],companysales['moisturizer'],labels=['Face Cream', 'Face Wash','toothpaste','bathingsoap', 'shampoo', 'moisturizer'])
plt.xticks(np.arange(0,13,2))
plt.yticks(np.arange(0,35000,5000))
plt.legend(loc='upper left')
plt.ylabel("Month Number")
```
